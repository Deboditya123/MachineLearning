```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import scipy.stats as stats
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score

from sklearn.preprocessing import PowerTransformer
```

```python
df=pd.read_csv('concrete_data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
x=df.drop(columns=['Strength'])
y=df.iloc[:,-1]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
lr=LinearRegression()

lr.fit(X_train,Y_train)
y_pred=lr.predict(X_test)

r2_score(Y_test,y_pred)
```

```python
np.mean(cross_val_score(lr,x,y,scoring='r2'))
```

```python
for col in X_train.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(X_train[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(X_train[col],dist='norm',plot=plt)
    plt.title(col)

    plt.show()
```

```python
pt= PowerTransformer(method='box-cox')

X_train_transformed=pt.fit_transform(X_train+0.00000001)
X_test_transformed=pt.fit_transform(X_test+0.00000001)

pd.DataFrame({'cols':X_train.columns,'box_cox_lambdas':pt.lambdas_})
```

```python
lr=LinearRegression()

lr.fit(X_train_transformed,Y_train)
y_pred=lr.predict(X_test_transformed) 

r2_score(Y_test,y_pred)
```

```python
X_transformed=pt.fit_transform(x+0.0000000001)
```

```python
np.mean(cross_val_score(lr,X_transformed,y,scoring='r2'))
```

```python
X_train_transformed_d=pd.DataFrame(X_train_transformed,columns=X_train.columns)

for col in X_train_transformed_d.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(X_train[col])
    plt.title(col)

    plt.subplot(122)
    sns.distplot(X_train_transformed_d[col])

    plt.title(col)

    plt.show()
```

```python
pt= PowerTransformer()

X_train_transformed2=pt.fit_transform(X_train)
X_test_transformed2=pt.fit_transform(X_test)

pd.DataFrame({'cols':X_train.columns,'Yeo_Johnson_lambdas':pt.lambdas_})
```

```python
lr.fit(X_train_transformed2,Y_train)

y_pred3=lr.predict(X_test_transformed2)

print(r2_score(Y_test,y_pred3))
```
